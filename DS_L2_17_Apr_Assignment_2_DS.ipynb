### Visualize the ```df``` by drawing the following plots:

### Visualize the ```df``` by following the steps given below:

* Get a series containing counts of unique values of ```Attrition```.
* Draw a countplot for ```Attrition``` using ```sns.countplot()```. 

````
# Get a series of counts of values of Attrition
occurance_count = {}
for i in df["Attrition"]:
  if i not in occurance_count:
    occurance_count[i]=1
  else:
    occurance_count[i] +=1
print(occurance_count)
data = pd.DataFrame(list(occurance_count.items()), columns=['Attrition_type', 'Occurance'])
data
# Draw a countplot for Attrition
sns.countplot(x="Attrition",data=df,order=[0,1])
plt.show()
sns.countplot(y="Attrition",data=df,order=[0,1])
plt.show()
# You may use countplot of seaborn for this 
sns.barplot(x="Attrition_type",y="Occurance",data=data,order=[0,1]);
plt.show()
````

### Visualize the ```df``` by following the steps given below:
