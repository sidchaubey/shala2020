```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.utils import resample
from sklearn.metrics import accuracy_score
import seaborn as sns

```

```
x = np.random.randint(1,100,1000)
#print(x)
# Configure Bootstrap
n_iters = 1000
n_size = int(len(x))
medians= list()
for i in range(n_iters):
  m = resample(x,n_samples=n_size)
  med = np.median(m)
  medians.append(med)
# Plot score
plt.hist(medians)
plt.show()
sns.distplot(medians,color=(0.3,0.1,0.1))
plt.show()
alpha = 0.2
p = ((1-alpha)/2.0)*100
lower = np.percentile(medians,p)

p = ((1+alpha)/2.0)*100
upper = np.percentile(medians,p)

print("The ci for {} percent lies between {} and {}".format(alpha*100,lower,upper))
```

```
# Central Limit Theorem

x = np.random.randint(1,1000,100)

n_iters = 1000
n_size = int(len(x))


means = list()
for i in range(n_iters):
  s = resample(x,n_samples=n_size)
  mean_ = np.mean(s)
  means.append(mean_)

plt.hist(means)# The graph looks like gaussian distribution
plt.show()
sns.distplot(means,color=(0.1,0.1,0.1))
plt.show()
```
